## Boolean arrays - A way to access a given set of elements in an array

This section provides some useful background that you may choose employ for exercise #6.2 below.

If you have an array, you can perform queries based on values of that array.  These queries return "Boolean arrays", which will contain True or False values depending on whether the element of the array meets the criteria.  You can then use these Boolean arrays to only access the value you would

```python
#this example shows you how to create and use a boolean array witha single argument
a = np.array([1,2,3,4,5])           #create an array
#perform a query.
#I always end the the output of a Boolean query out with "flag" so that I know it's a 
#Boolean array
asmallflag = a < 4           

#you can see that only the first three elements of asmall are "True"
print('A boolean array that selects small a values:', asmallflag)

#now using 'abig' as an array of indices for 'a' you can print out only the elements of a 
#that meet your query, that is, those that have True indices
print('Only the elemens of a for which the query yielded a True:', a[asmallflag])

#Now let's make a second array with the same length as a
b = np.array(['aa','bb','cc','dd','ee'])

# in this case, b could be names of object in a table and a could be some value associated 
#with those objects.  Since they have the same length, same order, **and** refer to the same set of data
#we can apply the 'abig' Boolean array to find out which elements of b are selected when we 
#perform our query on a
print('Only the elements of b that were selected by the previous query:', b[asmallflag])

#that's pretty cool!  But wait, there's more
#you can also make Boolean arrays by combining different conditions.  You can do this two different
#wasy.  
#First, you can set up a single query with two conditions, each enclosed in ()
amidflag = (a>2) & (a<4)
print('The elements of a and b that meet the new criteria: ', a[amidflag], b[amidflag])

#but I can also make individual Boolean arrays and combine them with a logical AND (&) command.
#Only elements that are True in both input arrays are true in the output array
abigflag = a>2
amidflag2 = abigflag & asmallflag
print('The results of combining the two previous selections:',amidflag2, a[amidflag2])

#and finally, we can assign only the values that satisfy a boolean command to a new arraw.
#in this case, only values of a with True indicies in amidflag2 will be assigned to c
onlymidvalues = a[amidflag2]
print(onlymidvalues)

#in the exercise below, we will use this technique to print out the flux at a point
#on the spectrum with a given wavelength
```

```python
#write your code here

# Define the wavelength ranges for the g-band and r-band in angstroms
lamgarr = 5150  # g-band wavelength in angstroms
lamrarr = 6350  # r-band wavelength in angstroms

# The 'star1' and 'star3' lambda arrays are already in angstroms
# These arrays represent the wavelengths at which the flux values (flam) are given
star1_lam_m = star1['lambda']  # Wavelength for star1 in angstroms
star3_lam_m = star3['lambda']  # Wavelength for star3 in angstroms

# Calculate the frequencies (nu) using the speed of light (c = 3e8 m/s)

# This converts from angstroms to meters before computing the frequency
star1nu = (3*10**8) / (star1_lam_m * 10**-10)  # Frequency for star1 in Hz
star3nu = (3*10**8) / (star3_lam_m * 10**-10)  # Frequency for star3 in Hz

# Define a function to convert flux per unit wavelength (f_lambda) to flux per unit frequency (f_nu)
# Flux per unit wavelength is usually given in erg/s/cm²/Å, while flux per unit frequency is erg/s/cm²/Hz
def convert_lambda_to_nu_flux(lambda_array, flam_array):
    c = 3e8  # speed of light in m/s
    lambda_m = lambda_array  # Wavelength in angstroms, used as is for conversion
    f_nu = (flam_array * 10**-10) * (lambda_m**2) / c  # Flux per unit frequency (erg/s/cm²/Hz)
    return f_nu

# Get the flux per unit wavelength (f_lambda) at the relevant wavelengths (g-band and r-band)
# We are looking for the closest wavelength indices for the g-band and r-band
# Use np.abs() to find the closest wavelengths to the target g-band (4800) and r-band (6100)

flam_gband_star1 = star1['flam'][np.abs(star1['lambda'] - lamgarr).argmin()]  # Flux at g-band for star1
flam_rband_star1 = star1['flam'][np.abs(star1['lambda'] - lamrarr).argmin()]  # Flux at r-band for star1

flam_gband_star3 = star3['flam'][np.abs(star3['lambda'] - lamgarr).argmin()]  # Flux at g-band for star3
flam_rband_star3 = star3['flam'][np.abs(star3['lambda'] - lamrarr).argmin()]  # Flux at r-band for star3

# Convert flux per unit wavelength (f_lambda) to flux per unit frequency (f_nu) for both stars
fnu_gband_star1 = convert_lambda_to_nu_flux(lamgarr, flam_gband_star1)  # Flux at g-band for star1 in erg/s/cm²/Hz
fnu_rband_star1 = convert_lambda_to_nu_flux(lamrarr, flam_rband_star1)  # Flux at r-band for star1 in erg/s/cm²/Hz

fnu_gband_star3 = convert_lambda_to_nu_flux(lamgarr, flam_gband_star3)  # Flux at g-band for star3 in erg/s/cm²/Hz
fnu_rband_star3 = convert_lambda_to_nu_flux(lamrarr, flam_rband_star3)  # Flux at r-band for star3 in erg/s/cm²/Hz

# Print the flux density per unit frequency (f_nu) values for both stars at g-band and r-band
print(f'fnu of star1 at the g-band = {fnu_gband_star1:4.2e}')
print(f'fnu of star1 at the r-band = {fnu_rband_star1:4.2e}')
print(f'fnu of star3 at the g-band = {fnu_gband_star3:4.2e}')
print(f'fnu of star3 at the r-band = {fnu_rband_star3:4.2e}')

# The AB magnitude formula: AB = -2.5 * log10(f_nu) - 48.6
# Using the calculated fnu values to compute the AB magnitudes
ABmag_star1_gband = -2.5 * np.log10(fnu_gband_star1) - 48.6
ABmag_star1_rband = -2.5 * np.log10(fnu_rband_star1) - 48.6
ABmag_star3_gband = -2.5 * np.log10(fnu_gband_star3) - 48.6
ABmag_star3_rband = -2.5 * np.log10(fnu_rband_star3) - 48.6

# Print the AB magnitudes for both stars at g-band and r-band
print(f'AB of star1 at the g-band = {ABmag_star1_gband:4.2f}')
print(f'AB of star1 at the r-band = {ABmag_star1_rband:4.2f}')
print(f'AB of star3 at the g-band = {ABmag_star3_gband:4.2f}')
print(f'AB of star3 at the r-band = {ABmag_star3_rband:4.2f}')

# Calculate the g-r color index for each star
# The g-r color is the difference between the AB magnitudes in the g-band and r-band
gr_star1 = ABmag_star1_gband - ABmag_star1_rband  # g-r color for star1
gr_star3 = ABmag_star3_gband - ABmag_star3_rband  # g-r color for star3

# Print the g-r color values for both stars
print(f'g-r color of star1 = {gr_star1:1.2f}')
print(f'g-r color of star3 = {gr_star3:1.2f}')
```
